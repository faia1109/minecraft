#マイクラのコマンドでよく使うもの




---


##コマンドの使い方
コマンドは**コメントライン**と**コマンドブロック**にコードを書き込むことで実行することができます。
基本的には

```
/<コマンドの内容>
```
という書き方になっています。

ここから実行例とともによくコマンドで使うものを紹介していきます。





---

##ゲームモードコマンド

いつも設定でゲームモードを変えるのは面倒くさくないですか？

変えることができるようになるのがこのコマンド


このコマンドは基本的には以下のような構文です


```
/gamemode <ゲームモード> <対象>
```
<ゲームモード>は以下の４つです
* survival ,s ,0  :サバイバルモード
* creative ,c ,1  :クリエイティブモード
* adventure ,a ,2 :アドベンチャーモード
* default ,d ,5   :デフォルトのゲームモード‌
* spectator       :スペクテイターモード(見る専)

<対象>は<ターゲット>で選択できます。省略すると実行者になります。

プレイヤー以外を選択するとエラーになります。

以下は実行例です。



```
# 一番近くのプレイヤーをクリエイティブモードにする
/gamemode 1 @p

# "実行者"をスペクテイターモードにする
/gamemode spectator

#すべてのプレイヤーをサバイバルモードにする
/gamemode 0 @a
```

##テレポートコマンド
このコマンドはコメントラインの部分にボタンが設定してあり使ったこともあると思います。
もう少し深堀するともっと便利に使用することができます。

このコマンドは基本的には以下のような構文です  


```
/tp <テレポート対象> <行先>
```
<br>
<テレポート対象>はその名の通りテレポートさせる対象を書き込みます。

プレイヤーや、アイテムなど様々なものを<ターゲット>を通じて選択できます。

<行先>は<ターゲット>で選択した対象、または座標を書き込むことができます。(おそらく単体専用)

以下は実行例です。
```
#翔汰を-32,40,56にテレポートさせる
/tp shouta -32 40 56

#一番近いプレイヤーを翔汰にテレポートさせる
/tp @p shouta

#すべてのプレイヤーを翔汰にテレポートさせる
/tp @a shouta
```

##ギブコマンド
冒頭に言ったコマンドブロックなど特殊なブロックはクリエイティブでも出すことができません。

そういった特殊なものから普通の肉などすべてのアイテムを手に入れることができるコマンドです。

このコマンドは基本的には以下のような構文です。


```
/give <対象> <アイテムID> <個数>
```
<対象>は<ターゲット>を通じて選択できます。

<アイテムID>は様々なものがあります。コマンドでしか入手できないアイテムをいくつか紹介します。
* command_block : コマンドブロック
* barrier : バリアブロック(見えないブロック)
* structure_block : ストラクチャーブロック
* structure_void : ストラクチャーヴォイド

<個数>は個数です。省略すると1つ手に入ります。自分が持てる量を超えると<対象>の目の前に巻き散らかされます

以下は具体例です


```
#すべてのプレイヤーにベットを1つ与える
/give @a bed

#一番近くのプレイヤーに松明を64つ与える
/give @p torch

#翔汰にステーキを1000つ与える
/give shouta cooked_beef 1000
```

##クリアコマンド

持ち物がいっぱいの時、一つづつ捨てるのは面倒くさいですよね。

そういう時に使えるのがこのコマンドです。

対象の持ち物すべてを消すことができます。


このコマンドは基本的には以下のような構文です

```
# 構文
/clear <対象>
```
<対象>は<ターゲット>を通じて選択できます

以下が具体例です
```
#一番近くのプレイヤーの持ち物をすべて消す
/clear @p

#すべてのプレイヤーの持ち物をすべて消す
/clear @a
```
> ・注意<br>
装備品などすべてのアイテムも消えるので、消したくないものはいったんチェストに入れるか、一旦そこらへんにポイ捨てしましょう

##エフェクトコマンド

長い時間ポーション効果が欲しいとき、強いポーション効果が欲しい時ありますよね。

よく、暗視効果を付けているのがこのコマンドになっています。

ポーション効果を付けることができます。


このコマンドは基本的には以下のような構文です


```
/effect <対象> <エフェクトID> <効果時間> <強さ>
```
<対象>は<ターゲット>を通じて選択できます。

<エフェクトID>は様々なものがあるので、よく私が使うものを紹介します。
* night_vision:暗視
* speed:移動速度上昇
* water_breathing:水中呼吸
* clear:エフェクト効果の全消去(<効果時間><強さ>は記載しない)

<効果時間>は0~1,000,000まで設定できます。省略すると30秒になります。

<強さ>は0~225まで設定できます。省略すると0(最低の効果)になります。

以下が具体例です


```
# すべてのプレイヤーに暗視効果を百万秒つける
/effect @a night_vision 1000000

#すべてのゾンビに移動速度上昇LV225を百万秒つける
/effect @e[type=zombie] speed 1000000 225

#すべてのエンティティのエフェクト効果を消す
/effect @e clear
```

##召喚コマンド

エンダードラゴンを大量に呼びたいときありますよね。

そんな時に使えるコマンドがこのコマンドです。

様々なエンティティを召喚することができます。


このコマンドは基本的には以下のような構文です


```
/summon <エンティティID> <座標>
/summon <エンティティID> <名前> <座標>
```
<エンティティID>は様々なものがあります。いくつか挙げますが様々なものがあるので別の資料を参照してください。
* NPC:NPC。それ以上でも以下でもない
* zombie:ゾンビ
* ender_dragon:エンダードラゴン
* snow_golem:スノーゴーレム
* iron_golem:アイアンゴーレム

<座標>は座標。それ以上でも以下でもない。省略すると、"実行者"のいる場所になる

<名前>はそのエンティティの名前。名札で名前つけるのと同じ状況になる。名前は「"」で囲む必要がある

以下が具体例です。


```
# アイアンゴーレムを"実行者"のいる場所に召喚する
/summon iron_golem

#「翔汰のおばあちゃん」という名前の付いたNPCを1,42,-1475に召喚する
/summon NPC "翔汰のおばあちゃん" 1 42 -1475

#エンダードラゴンを59,8,75に召喚する
/summon ender_dragon 59 8 75
```

対象をキルすることができます。

このコマンドは基本的には以下のような構文です。

```
# すべてのアイテムをキルする
/kill @e[type=item]

#すべてのプレイヤーをキルする
/kill @a
```

##埋め立てコマンド

一定以内の範囲を石で埋めたいときありますよね。

そんな時はこのコマンドです。

様々なモードで埋め立てることができます。

このコマンドは基本的には以下のような構文です。


```
/fill <座標> <座標> <ブロックID> <モード>
/fill <座標> <座標> <ブロックID_1> replace <ブロックID_2>
```
<座標>は座標。それ以上でも以下でもない。

<ブロックID>は様々なものがあります。いくつか挙げますが様々なものがあるので別の資料を参照してください。
* air : 空気
* stone : 石
* grass_block : 草ブロック
* dirt : 土
* cobblestone : 丸石
* oak_planks : オークの木材
* sand : 砂
* glass : ガラス

オークの原木や、階段など方向や状態がさまざまあるものは<ブロックID>の後ろに[ ]を付けて指定することができます

<モード>は以下の6つです
* (省略) : すべてを指定のブロックで埋める
* destroy : 指定した範囲にもともとあるブロックをアイテム化し(適正ツールで破壊した時と同等)、すべてを指定のブロックで埋める
* hollow : 指定した範囲の外側だけ指定のブロックで埋め、中を空気で埋める
* keep : 指定した範囲の空気の部分だけ指定のブロックで埋める
* outline : 指定した範囲の外側だけ指定のブロックで埋め、中はそのまま
* replace : 指定した範囲の<ブロックID_2>を<ブロックID_1>に置き換える。<ブロックID_2>を省力すると<(省略)モード>と同等になる

以下が実行例です


```
#54,84,26から74,80,46の範囲を空気で埋める(消す)
/fill 54 84 26 74 80 46 air

#106,20,1800から120,40,1810の範囲の水をガラスに置き換える
/fill 106 20 1800 120 40 1810 glass replace water

#実行者から、x,z座標に+9マスの範囲の空気の部分を上付きのオークのハーフブロックで埋める
/fill ~ ~ ~ ~9 ~ ~9 wooden_slab[“minecraft:vertical_half”=”top”] keep
```
>注意<br>
領域には上限32768ブロックが指定されています。もし、それ以上の範囲を埋め立てる場合は分割してやりましょう



##クローンコマンド
建築物を移動したり、増やしたりしたいときありますよね。

そんな時にはこのコマンドを使います。

このコマンドを使うとコピペができます。

最近ではストラクチャーブロックという感覚的にコピペできるブロックが追加されたので、そちらでするのも有り。

このコマンドは基本的には以下のような構文です。


```
/clone <コピー元座標> <コピー元座標> <ペースト先座標> <マスクモード> <クローンモード>
/clone <コピー元座標> <コピー元座標> <ペースト先座標> filtered <クローンモード> <ブロックID>
```
<コピー元座標>は座標。それ以上でも以下でもない。

<ペースト先座標>は座標。**x,z座標の一番小さいところ**を中心にペーストするので注意が必要

<マスクモード>はどのようにしてコピーするかの指定
* filtered : 指定した範囲の中で指定したブロックのみ<ペースト先座標>に張り付ける。
* masked : 指定した範囲の中で空気以外のブロックを<ペースト先座標>に張り付ける。(空気の部分はそのままになるということ)
* replace : 指定した範囲の全ブロックをコピー先に張り付ける
* (省略) : replaceと同等。ただし、クローンモードは指定できないはず。

<クローンモード>はコピー先の処理を指定
* force : コピー元とペースト先が重なった場合、強制的にペーストする
* move : ペーストすると、コピー元の対象となったブロックがなくなる
* normal : 通常のコピーをする。コピー元とペースト先が重なった場合エラーが発生する。
* (省略) : normalと同等

以下が実行例です。


```
#10,-10,10から12,-8,12のすべてを今いる位置にコピペする
/clone 10 -10 10 12 -8 12 ~ ~ ~  

#100,10,100から80,10,80の範囲内の水をを60,8,60に移動させる
/clone 100 10 100 80 10 80 60 8 60 filtered move water
```

> ターゲットについて

ターゲットにはユーザーネームを直接設定するもののほかに、すべてのプレイヤーや、アイテムなどの様々な条件を設定できる<ターゲットセレクター>という機能があります。

ターゲットセレクターには以下のものがあります
*   @a : すべてのプレイヤー
*   @r : ランダムなプレイヤー
*   @e : すべてのエンティティ(プレイヤー,アイテム,モブ,着火されたTNTなど動くもの)
*   @p : 最も近いプレイヤー(コメントラインで実行すると自分自身)
*   @s : 自分自身(コマンドを実行したエンティティ)

<br>
また、ターゲットセレクターの後ろに[ ]を付けることで条件を付けることができます。

非常に多くの条件があるので、よく私が使うものを紹介します。
*   r=<距離> : <距離>に具体的な整数を入れることで、半径？マス以内のものに限定できます。
*   type=<対象> : <対象>に具体的な対象のIDを入れることで、対象を絞ることができます(例:item,zombie,sheepなど)
*   name=<名前> : <名前>に名前を入れることで、対象についている名前で絞ることができます。よく、[type=item,name="卵"]のようにして、対象を絞ることがあります。

以下はターゲットセレクターを使用した実行例です

```
#半径30マス以内のすべてのプレイヤーをランダムなプレイヤーの場所にテレポートさせる
/tp @a[r=30] @r
#すべてのアイテムを殺す
/kill @e[type=item]
#すべての卵という名前の付いたアイテムを最も近いプレイヤーの場所にテレポートさせる
/tp @e[type=item,name="卵"] @p
```

>座標について

<座標>は既知の通りx,y,z座標をもとに構成されています。

固定の場所でやるのであれば座標を打ち込んでやってもいいのですが、今いる場所から10マス足したところにテレポートしたいと思ったことはないでしょうか?

こういう時に役に立つのが相対座標です。

相対座標に使うマークが「**~**」です。

「~」は現在いる位置という意味で使われ、その後ろに数値を書くことで？マス離れた場所といった書き方ができるのです。

以下は具体例です


```
# 一番近くのプレイヤーを"実行者"からy軸に+200マス上にテレポートさせる
/tp @p ~ ~200 ~
#現在いる場所から-20,5,40まで空気で埋める(消す)
/fill ~ ~ ~ -20 5 40 air
```
* 注意

マイクラには演算距離というものがあります。

これはワールド設定で変えれるのですが、ブロックの更新などをする場合はこの範囲内に収めなければエラーになります。

演算距離を確認しながらそれ以上の範囲を扱う場合には分割をしましょう

>演算距離で使用されるチャンクとは16x16の柱状の範囲のことです
